# Simple Linear Regresssion

```python
#Y=mx+c
```

```python
class LR:
    def __init__(self):
        self.m=None
        self.c=None

    def fit(self, X_train, y_train):

        num=0
        den=0
        for i in range(len(X_train)):
            num=num+(X_train.iloc[i]-X_train.mean())*(y_train.iloc[i]-y_train.mean())
            den=den+(X_train.iloc[i]-X_train.mean())*(X_train.iloc[i]-X_train.mean())


        self.m=num/den
        self.c=y_train.mean()-self.m*X_train.mean()


    def predict(self, X_test):

        return self.m*X_test + self.c
```

# Multiple Linear Regression

```python
class MLR:
    def __init__(self):
        self.coef=None
        self.intercept=None

    def fit(self, X_train, y_train):
        X_train=np.insert(X_train, 0, 1, axis=1)
        betas=np.linalg.pinv(X_train.T @X_train) @X_train.T @ y_train

        self.intercept=betas[0]
        self.coef=betas[1:]

    def predict(self, X_test):
        y_pred =np.dot(X_test, self.coef) + self.intercept

        return y_pred
        
```
